```python
import numpy as np

from scipy.io import loadmat

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
np.set_printoptions(linewidth=1000)
```

```python
_binary_result = loadmat('./binary/result_Gen-CUDE.mat')
_4ary_result = loadmat('./4ary/result_Gen-CUDE.mat')
_10ary_result = loadmat('./10ary/result_Gen-CUDE.mat')
```

```python
_binary_k_set = _binary_result['window_size_k'][0]
_binary_Gen_CUDE_error_unnormalized = _binary_result['error_rate'][0]
_binary_Gen_CUDE_time = _binary_result['time'][0]
_binary_error_x_z = _binary_result['error_x_z'][0]
_binary_Gen_CUDE_error = _binary_Gen_CUDE_error_unnormalized/_binary_error_x_z
print(_binary_Gen_CUDE_error)

_4ary_k_set = _4ary_result['window_size_k'][0]
_4ary_Gen_CUDE_error_unnormalized = _4ary_result['error_rate'][0]
_4ary_Gen_CUDE_time = _4ary_result['time'][0]
_4ary_error_x_z = _4ary_result['error_x_z'][0]
_4ary_Gen_CUDE_error = _4ary_Gen_CUDE_error_unnormalized/_4ary_error_x_z
print(_4ary_Gen_CUDE_error)

_10ary_k_set = _10ary_result['window_size_k'][0]
_10ary_Gen_CUDE_error_unnormalized = _10ary_result['error_rate'][0]
_10ary_Gen_CUDE_time = _10ary_result['time'][0]
_10ary_error_x_z = _10ary_result['error_x_z'][0]
_10ary_Gen_CUDE_error = _10ary_Gen_CUDE_error_unnormalized/_10ary_error_x_z
print(_10ary_Gen_CUDE_error)
```

```python
plt.rcParams["figure.figsize"] = (20,5)
plt.rcParams['lines.linewidth'] = 2
plt.rcParams['axes.grid'] = False 
plt.rcParams['font.size'] = 15


########################################################################################
ax1=plt.subplot(131)


plt.plot(_binary_k_set, _binary_Gen_CUDE_error, color='BLUE',marker='s',linestyle='solid',label = 'Gen-CUDE')

plt.xticks(_binary_k_set)
plt.ylim(-0.05,1.05)
plt.yticks(np.arange(0,1.10,0.1))

plt.title("$|\mathcal{A}|=2$")
plt.xlabel('Window Size K')
plt.ylabel("Normalized Error Rate")


########################################################################################
ax2=plt.subplot(132)

plt.plot(_4ary_k_set, _4ary_Gen_CUDE_error, color='BLUE',marker='s',linestyle='solid',label = 'Gen-CUDE')

plt.xticks(_4ary_k_set)
plt.ylim(-0.05,1.05)
plt.yticks(np.arange(0,1.10,0.1))

plt.title("$|\mathcal{A}|=4$")
plt.xlabel('Window Size K')
plt.ylabel("Normalized Error Rate")

########################################################################################

ax3=plt.subplot(133)

plt.plot(_10ary_k_set, _10ary_Gen_CUDE_error, color='BLUE',marker='s',linestyle='solid',label = 'Gen-CUDE')

plt.xticks(_10ary_k_set)

plt.ylim(-0.05,1.05)
plt.yticks(np.arange(0,1.10,0.1))

plt.title("$|\mathcal{A}|=10$")
plt.xlabel('Window Size K')
plt.ylabel("Normalized Error Rate")
handles, labels = ax3.get_legend_handles_labels()

lgd = ax3.legend(handles, labels, loc='right', bbox_to_anchor=(0,0,1.63,1),ncol=1, mode="None", borderaxespad=0)


plt.savefig('error_rate-plots.png', bbox_extra_artists=(lgd,), bbox_inches='tight',dpi=100)
plt.show()
```
